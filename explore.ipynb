```python
import os, glob
csv_files = glob.glob("./metadata/*.csv")
```

```python
inchi_to_smiles = {}
for file in csv_files:
    print(f'reading {file} ...')
    df = pd.read_csv(file)
    for index, row in df.iterrows():
        inchi_to_smiles[row['INCHIKEY']] = row['SMILES']
```

```python
zero_count = (filtered_data['INCHIKEY'].apply(lambda x: inchi_to_smiles.get(x,0)) == 0).sum()
zero_count
```

```python
filtered_data['SMILES'] = filtered_data['INCHIKEY'].apply(lambda x: inchi_to_smiles.get(x,0))
```

```python
columns = list(filtered_data.columns)
columns.insert(2,columns.pop(columns.index('SMILES')))
filtered_data = filtered_data[columns]
filtered_data.head()
```

```python
filtered_data.to_pickle('./filtered_image_embeddings.pkl')
```

```python
from sklearn.model_selection import train_test_split
def split_dataset(dataset, seed):
    unique_inchikeys = dataset.INCHIKEY.unique()
    train_keys, temp_keys = train_test_split(unique_inchikeys, test_size=0.3, random_state=seed)
    val_keys, test_keys = train_test_split(temp_keys, test_size=0.5, random_state=seed)

    train = dataset.loc[dataset.INCHIKEY.isin(train_keys)].reset_index(drop=True)
    val = dataset.loc[dataset.INCHIKEY.isin(val_keys)].reset_index(drop=True)
    test = dataset.loc[dataset.INCHIKEY.isin(test_keys)].reset_index(drop=True)

    return train, val, test
```

```python
train_df, val_df, test_df = split_dataset(filtered_data, seed=42)
```

```python
train_df.to_pickle('./train_df.pkl')
val_df.to_pickle('./val_df.pkl')
test_df.to_pickle('./test_df.pkl')
```
